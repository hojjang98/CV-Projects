```python
import numpy as np
import matplotlib.pyplot as plt

# OR gate dataset
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([[0], [1], [1], [1]])

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    s = sigmoid(x)
    return s * (1 - s)

# Mean squared error loss
def mse(pred, target):
    return np.mean((pred - target) ** 2)

def mse_grad(pred, target):
    return 2 * (pred - target) / target.size

# Weight initialization
np.random.seed(42)
W1 = np.random.randn(2, 4)
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 1)
b2 = np.zeros((1, 1))

lr = 0.1
losses = []

# Training loop
for epoch in range(5000):
    # Forward pass
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)

    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)

    loss = mse(a2, y)
    losses.append(loss)

    # Backward pass
    dL = mse_grad(a2, y)
    dz2 = dL * sigmoid_deriv(z2)
    dW2 = a1.T @ dz2
    db2 = dz2.sum(axis=0, keepdims=True)

    dz1 = dz2 @ W2.T * sigmoid_deriv(z1)
    dW1 = X.T @ dz1
    db1 = dz1.sum(axis=0, keepdims=True)

    # Weight update
    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2

    if epoch % 500 == 0:
        print(f"Epoch {epoch} | Loss: {loss:.4f}")

# Final predictions
print("\nFinal Predictions:")
for i in range(len(X)):
    z1 = X[i:i+1] @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    pred = sigmoid(z2)
    print(f"Input: {X[i]} → Pred: {pred[0][0]:.4f} | Target: {y[i][0]}")

# Plot loss curve
plt.plot(losses)
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid()
plt.show()
```
